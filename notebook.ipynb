```python
num_cols = ['temperature_10H_max (°C)', 'temperature_10H_min (°C)', 'temperature-1', 'temperature-2', 'temperature-3', 'apparent_temperature_max', 'apparent_temperature_min', 'humidity', 'Voltage-L', 'Voltage-R', 'Voltage-M', 'Current-M', 'Current-R', 'Current-T', 'RPM', 'RPM-1', 'RPM-2', 'RPM-3', 'Vibration-1', 'Vibration-2', 'Power'] 
train_df[num_cols] = train_df[num_cols].fillna(train_df[num_cols].mean())

cat_cols = ['ID_Area', 'Machine', 'Power_Backup'] 
train_df[cat_cols] = train_df[cat_cols].apply(lambda x: x.fillna(x.mode()[0]))
```
